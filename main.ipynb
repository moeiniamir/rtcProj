```python
import task_generation.taskset_generation as tg
import numpy as np
from itertools import chain, repeat
```

```python
u = np.array(tg.generate_uunifastdiscard(10, .5, 100, 'test.csv'))
u
```

```python
available_periods = list(range(1, 11))
p = np.array(tg.generate_random_periods_discrete(100, 10, available_periods))
p
```

```python
tset = np.array(tg.generate_tasksets(u, p, 'tset.csv'), dtype=[('cost', 'float64'), ('period', 'int32')])
tset.shape
```

```python
# get the hyper period of the task set
hp = np.lcm.reduce(tset['period'], axis=1)
hp
```

```python
# get release times, costs and deadlines for each taskset based on period and cost for a given hyper period
tasks = [
    list(chain.from_iterable([
        zip(
            list(range(0, hp[i_tset], tset['period'][i_tset, i_task])),
            list(range(tset['period'][i_tset, i_task], hp[i_tset] + 1, tset['period'][i_tset, i_task])),
            list(repeat(tset['cost'][i_tset, i_task], hp[i_tset] // tset['period'][i_tset, i_task]))
        ) for i_task in range(len(tset[i_tset]))
    ]))
    for i_tset in range(len(tset))
]
```

```python
for i, taskset in enumerate(tasks):
    print(i, len(taskset))
```
